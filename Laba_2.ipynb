```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

1.Загрузите выборку из файла

```python
data = pd.read_csv('/content/sample_data/Titanic-Dataset.csv', index_col="PassengerId")
data
```

2. Оставьте в выборке 4 параметра

```python
X = data[['Pclass', 'Fare', 'Age', 'Sex', 'Survived']]
X.head()
```

3. Sex имеет строковые значения

```python
X.loc[X['Sex'] == 'male'] = 0
X.loc[X['Sex'] == 'female'] = 1
X.head()
```

4. Выделение целевой переменной

```python
Y = X['Survived']
print(Y.head())
```

5. Удаление пропущенных значений

```python
X = X[X[['Pclass', 'Fare', 'Age', 'Sex', 'Survived']].notna()]
X.head()
```

6. Обучение решающего дерева

```python
clf = DecisionTreeClassifier(random_state=241)
clf.fit(X, Y)
```

7. Вычисление важности признака

```python
importances = clf.feature_importances_

feature_importances_df = pd.DataFrame({'Feature': X.columns, 'Importance': importances})
feature_importances_df = feature_importances_df.sort_values(by='Importance', ascending=False)

print("\nВажность признаков:")
print(feature_importances_df)

top_2_features = feature_importances_df['Feature'].head(2).tolist()
answer = ",".join(top_2_features)

print(f"{answer}")
```
