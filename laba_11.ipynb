```python
import pandas as pd
from sklearn.decomposition import PCA
from numpy import corrcoef
```

1. Загрузите выборку

```python
data = pd.read_csv('/content/close_prices.csv')
data
```

2. Обучите преобразование PCA

```python
pca = PCA(n_components=10)
result = pca.fit(data.iloc[:,1:31])
variance_ratio = result.explained_variance_ratio_
components = result.components_
```

```python
componentsCount = 1
while sum(variance_ratio[:componentsCount])<0.9:
  componentsCount += 1

componentsCount
```

3. Примите к исходным преобразованиям

```python
result = pca.transform(data.iloc[:,1:31])
firstComponent = result[:,0]
result
```
